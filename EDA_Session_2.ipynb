**Create data frame using dictionary**

```python
name=['Avinash','Aakash','Aadhya']
age=[25,30,35]
pd.DataFrame(zip(name,age),columns=['Names','Ages'])
```

```python
dict1={'Names':['Avinash','Aakash','Aadhya'],
      'Ages':[25,30,35]}
dict1
```

```python
pd.DataFrame(dict1)
# Keys becomes columns
# Values becomes rows
# No need to provide columns names separately
```
